```python
import pandas as pd
import numpy as np
import time
```

```python
df_test = pd.read_csv("testing_dic.csv")
df_test[0:5]
```

```python
listerned = df_test.drop(['song_id','target','set','msno_int'], axis=1)
listerned[0:5]
```

```python
test = df_test.drop(['song_id','target','set','msno_int','song_id_int'],axis=1)
test[0:5]
```

```python
test.shape
```

```python
test = test.drop_duplicates()
test.shape
```

```python
test[0:5]
```

```python
song_pool = pd.read_csv("ALS.csv", header=None, names=['target', 'set','msno_int','song_id_int'])
song_pool[0:5]
```

```python
song_pool['w'] =song_pool.target+1
song_pool[0:5]
```

```python
song_pool.shape
```

```python
song_pool = song_pool.loc[song_pool.set==1]
song_pool.shape
```

```python
RLT = pd.DataFrame(index=range(test.shape[0]),columns=range(11))
RLT.columns = ['ID','R1','R2','R3','R4','R5','R6','R7','R8','R9','R10']
RLT.ID = list(test.msno)
RLT
```

```python
song_pool3={}
for i in range(song_pool.shape[0]):
    new_song = song_pool.iloc[i,3]
    if new_song in song_pool3:
        song_pool3[new_song]=song_pool3.get(new_song)+song_pool.iloc[i,4]
    else:
        song_pool3[new_song]=song_pool.iloc[i,4]
    if i%100000==0:
        print(i)
```

```python
for u in range(test.shape[0]):
    listerned_u=listerned.loc[listerned.msno==test.iloc[0,0]].song_id_int
    song_pool_u2=song_pool3
    for x in listerned_u:
        if x in song_pool_u2:
            song_pool_u2.pop(x)
    song_pool_u = list(song_pool_u2.keys())
    dist = list(song_pool_u2.values())
    dist=pd.Series(dist)/sum(dist)
    recommendation = np.random.choice(song_pool_u,10,False,dist)
    for r in range(10):
        RLT.iloc[u,r+1]=recommendation[r]
    if u%20==0:
        print(u)
```

```python
RLT.to_csv("RandomRltintID.csv", index=False)
```
